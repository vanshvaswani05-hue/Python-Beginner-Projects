```python
import random

def number_guessing_game():
    computer_number = random.randint(1, 100)
    attempt = 0

    print("-- Welcome in Number Guessing Game --")

    while True:
        user_input = input("Enter a value (1-100) or type 'quit': ")

        if user_input.lower() == "quit":
            print("Game Quit! 👋")
            print(f"The correct number was {computer_number}")
            break

        user_number = int(user_input)
        attempt += 1

        if computer_number > user_number:
            print("You guessed a low number! ⬇️")
        elif computer_number < user_number:
            print("You guessed a high number! ⬆️")
        else:
            print("You guessed the right number! 🎉")
            print(f"You guessed this number in {attempt} attempts!")
            break

number_guessing_game()
```
